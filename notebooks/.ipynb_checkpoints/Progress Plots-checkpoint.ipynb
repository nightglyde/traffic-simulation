```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def createProgressPlot(density, dataset_suffix, num_rows=None, start_rows=0, ax=None):
    
    filename = "../simulation/results/compiled_results/progress_dataset_1x1_{}_{}.txt".format(
        density, dataset_suffix)
    
    try:
        df = pd.read_csv(filename, delimiter=" ", header=None)
    except:
        return

    df.columns = ["Time",
                  "Traffic Lights",
                  "Virtual Traffic Lights",
                  "My Traffic Controller",
                  "Modified MTC"]

    #df.drop(columns=["My Traffic Controller2"], inplace=True)
    
    if num_rows != None:
        df = df.iloc[start_rows:num_rows]
    
    if ax is None:
        ax = df.set_index("Time").plot()
    else:
        df.set_index("Time").plot(ax=ax)
    
    title = "Progress of Traffic Controller Simulations with\nTraffic Density of {} Cars per Minute".format(
        int(density))

    ax.set_title(title)
    ax.set_xlabel("Time (minutes)")
    ax.set_ylabel("Successful Cars")
    
    savefile = "progress_plot_1x1_{}_{}.png".format(density, dataset_suffix)
    plt.savefig(savefile)
```

```python
def progressTestSuite(dataset_suffix, num_rows=None, start_rows=0):
    
    fig, axes = plt.subplots(5, 3, figsize=(28, 30))
    
    for i in range(15):
        x = i % 3
        y = i // 3
        density = "{:03}".format((i+1)*10)
        createProgressPlot(density, dataset_suffix, num_rows, start_rows, axes[y][x])
        
```

```python
def createDurationPlot(turn_ratio, short_name, long_name, ax=None):
    
    filename = "../simulation/results/compiled_results/duration_{}_dataset_1x1_{}.txt".format(short_name,
                                                                                              turn_ratio)
    
    df = pd.read_csv(filename, delimiter=" ", header=None)

    df.columns = ["Traffic Density",
                  "Traffic Lights",
                  "Virtual Traffic Lights",
                  "My Traffic Controller",
                  "Modified MTC"]

    #df.drop(columns=["My Traffic Controller2"], inplace=True)

    df = df.iloc[:10]

    df.set_index("Traffic Density", inplace=True)

    if ax == None:
        ax = df.plot()
    else:
        df.plot(ax=ax)

    ax.set_title("{} Journey Duration vs Traffic Density".format(long_name))
    ax.set_xlabel("Traffic Density (cars per minute)")
    ax.set_ylabel("{} Journey Duration (seconds)".format(long_name))

    plt.savefig("duration_{}_plot_1x1_{}.png".format(short_name, turn_ratio))
```

```python
def createFailsPlot(turn_ratio, ax=None):
    
    filename = "../simulation/results/compiled_results/fails_dataset_1x1_{}.txt".format(turn_ratio)
    
    df = pd.read_csv(filename, delimiter=" ", header=None)

    df.columns = ["Traffic Density",
                  "Traffic Lights",
                  "Virtual Traffic Lights",
                  "My Traffic Controller",
                  "Modified MTC"]

    #df.drop(columns=["My Traffic Controller2"], inplace=True)

    #df = df.iloc[7:]

    df.set_index("Traffic Density", inplace=True)

    
    if ax == None:
        ax = df.plot(kind='bar')
    else:
        df.plot(kind='bar', ax=ax)

    ax.set_title("Failed Test Cases vs Traffic Density")
    ax.set_xlabel("Traffic Density (cars per minute)")
    ax.set_ylabel("Number of Fails")

    plt.savefig("fails_plot_1x1_{}.png".format(turn_ratio))
```

```python
def durationTestSuite(turn_ratio):
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    createDurationPlot(turn_ratio, "mean", "Mean", axes[0][0])
    createDurationPlot(turn_ratio, "max", "Maximum", axes[0][1])
    createDurationPlot(turn_ratio, "stdev", "Standard Deviation of", axes[1][0])
    createDurationPlot(turn_ratio, "min", "Minimum", axes[1][1])
    
    createFailsPlot(turn_ratio)
```
