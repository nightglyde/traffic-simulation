```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def createProgressPlot(density, dataset_suffix, num_rows=None, start_rows=0):
    
    filename = "../simulation/results/compiled_results/progress_dataset_1x1_{}_{}.txt".format(
        density, dataset_suffix)
    df = pd.read_csv(filename, delimiter=" ", header=None)

    df.columns = ["Time",
                  "Traffic Lights",
                  "Virtual Traffic Lights",
                  "My Traffic Controller2",
                  "My Traffic Controller"]

    df.drop(columns=["My Traffic Controller2"], inplace=True)
    
    if num_rows != None:
        df = df.iloc[start_rows:num_rows]
    
    ax = df.set_index("Time").plot()
    
    title = "Progress of Traffic Controller Simulations with\nTraffic Density of {} Cars per Minute".format(
        int(density))

    ax.set_title(title)
    ax.set_xlabel("Time (minutes)")
    ax.set_ylabel("Successful Cars")
    
    savefile = "progress_plot_1x1_{}_{}.png".format(density, dataset_suffix)
    plt.savefig(savefile)
```

```python
averages_df = pd.read_csv("../simulation/results/compiled_results/duration_mean_dataset_1x1_111.txt",
                          delimiter=" ", header=None)

averages_df.columns = ["Traffic Density",
                       "Traffic Lights",
                       "Virtual Traffic Lights",
                       "My Traffic Controller2",
                       "My Traffic Controller"]


averages_df.drop(columns=["My Traffic Controller2"], inplace=True)

averages_df = averages_df.iloc[:11]

averages_df.set_index("Traffic Density", inplace=True)

ax = averages_df.plot()

ax.set_title("Mean Journey Duration vs Traffic Density")
ax.set_xlabel("Traffic Density (cars per minute)")
ax.set_ylabel("Mean Journey Duration (seconds)")

plt.savefig("duration_mean_plot_1x1_111.png")
```
