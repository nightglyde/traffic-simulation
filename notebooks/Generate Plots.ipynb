```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("../simulation/results/compiled_results/progress_dataset_1x1_090_115A.txt",
                 delimiter=" ", header=None)

df.columns = ["Time",
              "Traffic Lights",
              "Virtual Traffic Lights",
              "My Traffic Controller",
              "Greedy Controller"]

ax = df.set_index("Time").plot()

ax.set_title("Progress of Traffic Controller Simulations with\nTraffic Density of 90 Cars per Minute")
ax.set_xlabel("Time (minutes)")
ax.set_ylabel("Successful Cars")

plt.savefig("progress_plot_1x1_090_115A.png")
```

```python
averages_df = pd.read_csv("../simulation/results/compiled_results/duration_mean_dataset_1x1_111.txt",
                          delimiter=" ", header=None)

averages_df.columns = ["Traffic Density",
                       "Traffic Lights",
                       "Virtual Traffic Lights",
                       "My Traffic Controller",
                       "Greedy Controller"]

averages_df = averages_df.set_index("Traffic Density")

ax2 = averages_df.plot()

ax2.set_title("Average Journey Duration vs Traffic Density")
ax2.set_xlabel("Traffic Density (cars per minute)")
ax2.set_ylabel("Average Journey Duration (seconds)")

plt.savefig("duration_mean_plot_1x1_111.png")
```

```python
maximums_df = pd.read_csv("../simulation/results/compiled_results/duration_max_dataset_1x1_111.txt",
                          delimiter=" ", header=None)

maximums_df.columns = ["Traffic Density",
                       "Traffic Lights",
                       "Virtual Traffic Lights",
                       "My Traffic Controller",
                       "Greedy Controller"]

maximums_df = maximums_df.set_index("Traffic Density")

ax3 = maximums_df.plot()

ax3.set_title("Maximum Journey Duration vs Traffic Density")
ax3.set_xlabel("Traffic Density (cars per minute)")
ax3.set_ylabel("Maximum Journey Duration (seconds)")

plt.savefig("duration_max_plot_1x1_111.png")
```

```python
fails_df = pd.read_csv("../simulation/results/compiled_results/fails_dataset_1x1_111.txt",
                          delimiter=" ", header=None)

fails_df.columns = ["Traffic Density",
                    "Traffic Lights",
                    "Virtual Traffic Lights",
                    "My Traffic Controller",
                    "Greedy Controller"]

fails_df = fails_df.set_index("Traffic Density")

ax3 = fails_df.plot()

ax3.set_title("Failed Test Cases vs Traffic Density")
ax3.set_xlabel("Traffic Density (cars per minute)")
ax3.set_ylabel("Fails")

plt.savefig("fails_plot_1x1_111.png")
```
